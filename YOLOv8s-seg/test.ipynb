```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

%matplotlib inline
```

```python
import numpy as np
from PIL import Image, ImageDraw
```

```python
model_1 = YOLO('/Users/21010992/Desktop/miem_v4/runs/segment/train29/weights/best.pt')
```

```python
# img = cv2.imread('test_png/66_112.png')
img = Image.open('test_png/66_112.png')
```

```python
def predict_png(model, img):
    results = model.predict(img)
    result = results[0]
    masks = result.masks
    if masks[0]:
        mask = masks[0].data[0].numpy()
        for i in range(1, len(masks)):
            mask += masks[i].data[0].numpy()
    mask_img = Image.fromarray(mask,"I")
    return mask_img
```

```python
predict_png(img)
```

```python
img = Image.open('/Users/21010992/Desktop/miem_v4/formatted_dataset/test/images/1_2.png')
img_labeled = Image.open('/Users/21010992/Desktop/miem_v4/dataset_sectors/black&white/1_2.png')
fig, ax = plt.subplots(ncols=3, figsize=(15,15))
ax[0].imshow(img)
ax[0].axis('off')
ax[0].set_title('Изображение', fontsize=12)
ax[1].imshow(predict_png(model_1, img))
ax[1].axis('off')
ax[1].set_title('Маска сегментации', fontsize=12)
ax[2].imshow(img_labeled)
ax[2].axis('off')
ax[2].set_title('Эталон', fontsize=12)
plt.show()
```
